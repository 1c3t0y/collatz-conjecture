# Collatz function Behaivour

# Stopping time

```python
stoping_time_ratio(3, collatz_function_short)
```

```python
values = range(2, 100000)
stopping = [stoping_time_ratio(value, collatz_function_short) for value in values]
plt.figure(figsize = (18,9))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.axhline(y = 6.143161, color = 'r', linestyle = '-')
plt.scatter(values, stopping)
```
