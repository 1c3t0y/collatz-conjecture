# Stopping time

```python
collatz3 = CollatzProblem(range(2,100),start = 'orbits_and_periods', f = collatz_function)
```

```python
collatz3.stopping_time()
```

```python
collatz3.stopping_time_ratios
```

```python
stopping = [collatz3.stopping_time_ratios[value] for value in collatz3.values]
plt.figure(figsize = (18,9))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.axhline(y = 6.143161, color = 'r', linestyle = '-')
plt.scatter(collatz3.values, stopping)
```

```python
values = range(2, 2**16)
collatz4 = CollatzProblem(values, start = 'orbits_and_periods', f = collatz_function)
stopping = [collatz4.stopping_time_ratios[value] for value in collatz4.values]
plt.figure(figsize = (18,9))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.axhline(y = 6.143161, color = 'r', linestyle = '-')
plt.scatter(collatz4.values, stopping)
```
