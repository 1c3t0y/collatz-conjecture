```python
# Adding collatz package to python path
import sys
sys.path.append('..')

# importing the dynamical package
from collatz.collatz import CollatzProblem
from collatz.functions import collatz_function_short, collatz_function, collatz_lines

from collatz.plot import plot_dots, plot_function

#other useful libraries
import numpy as np
import random

import matplotlib.pyplot as plt
```

```python
X = list(range(1,10))
collatz1 = CollatzProblem(X, start = 'orbit', f = collatz_lines)
collatz1.orbits
```

```python
ylim_min = min([collatz_lines(x) for x in range(1,10)]) - 1
ylim_max = max([collatz_lines(x) for x in range(1,10)]) + 1

plot_dots(collatz_lines, display_mode = 'show',  
        title = "Collatz pairs $(n, C(n))$", range = (1,10), num = 10, 
        figsize=(15,8), 
        xlim = (-1,12), ylim = (ylim_min, ylim_max))
```
