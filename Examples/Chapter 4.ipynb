### fixed points

```python
lines_fixed = [lines_fixed_points(n) for n in range(-50,51)]
```

```python
lines = DDS(lines_fixed, collatz_lines, iterations = 10, stop_iterations = 10, start = 'orbit')
lines.search_periodic_orbits()
```

```python
lines_fixed = lines.search_periodic_orbits().keys()
lines = DDS(lines_fixed, collatz_lines, iterations = 10, stop_iterations = 10, start = 'orbit')
lines.plot_orbits(function_name = 'C', display_mode = 'show', label_data = False, legend = True, markers = None, title ='', figsize = (12,6))
```

```python
lines.orbits_orbits
```

### Orbits around fixed points

```python
epsilon = 0.0001
around_fixed = []
for fixed in lines_fixed:
    around_fixed.extend(list(np.arange(fixed - epsilon,fixed + epsilon, 0.0001)))
```

```python
lines = DDS(lines_fixed, collatz_lines, iterations = 100, stop_iterations = 100, start = 'orbit')
lines.search_periodic_orbits()
```
